```python
import glob
import pandas as pd
```

```python
paths = glob.glob('장애인등록 장애유형별 성별 장애인수(201801_202205)/*.csv')
```

```python
len(paths)
```

```python
paths[0]
```

```python
df = pd.read_csv(paths[0], encoding = 'cp949')
df['통계년월'].values[1]
```

```python
val = df['통계년월'].values[1]
df = df[df['통계시도명'] == '서울특별시'].groupby('통계시군구명').sum()['등록장애인수'].reset_index()
df['년도'] = val
df
```

```python
for i in range(len(paths)):
    globals()['df'+str(i)] = pd.read_csv(paths[i], encoding = 'cp949')
    val = globals()['df'+str(i)]['통계년월'].values[1]
    globals()['df'+str(i)] = globals()['df'+str(i)][globals()['df'+str(i)]['통계시도명'] == '서울특별시'].groupby('통계시군구명').sum()['등록장애인수'].reset_index()
    globals()['df'+str(i)]['년도'] = val
```
