```python
title_list = []
date_list = []
for i in range(15):
    for j in range(1,26):
        r = globals()['soup'+str(j)]
        title = r.select('a.article')
        date = r.select('td.td_date')
        t_val = title[i].get_text("", strip=True)
        d_val = date[i].get_text("", strip=True)
        title_list.append(t_val)
        date_list.append(d_val)
title_list
date_list
```

```python
df_t = pd.DataFrame(title_list)
df_t= df_t.rename(columns={0:'title'})
df_d = pd.DataFrame(date_list)
df_d= df_d.rename(columns={0:'date'})
```

```python
df_t
```

```python
df = pd.concat([df_t, df_d], axis = 1)
df
```
