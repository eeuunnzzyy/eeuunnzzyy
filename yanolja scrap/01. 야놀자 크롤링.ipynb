```python
r_url = []
for i in range(len(bf)):
    bf_url_num = bf_url[i].split('/')[2]
    r_url_val =  f'https://domestic-order-site.yanolja.com/review/properties/{bf_url_num}/reviews'
    r_url.append(r_url_val)
print(r_url[0])
```

```python
len(r_url)
```

```python
#range()안에 가져올페이지수 입력
reviews = []
review_score = []
가져올페이지수 = 10
for i in range(가져올페이지수):
    get_t_v = requests.get(r_url[i])
    real_soup = BeautifulSoup(get_t_v.content, 'html.parser')
    reviews.append(real_soup)
    try : 
        review_score_val = reviews[i].select('span.total-score')[0].get_text("", strip=True)
        review_score.append(review_score_val)
    except :
        review_score.append('신규')
review_score
```

```python
real_soup=[]
review_comments = []
for j in range(가져올페이지수):
    digin = r_url[j]
    driver = webdriver.Chrome(executable_path='/Users/jeon-eunji/Documents/GitHub/TIL/크롤링/chromedriver')
    driver.get(digin)
    real_soup_val = BeautifulSoup(driver.page_source, 'html.parser')
    real_soup.append(real_soup_val)
    try : 
        real_soup_comment = []
        for k in range(len(real_soup[j].select('div.content-wrap > p'))):
            real_soup_comment_val = real_soup[j].select('div.content-wrap > p')[k].get_text("", strip=True).replace('\n','.').replace('\t','')
            real_soup_comment.append(real_soup_comment_val)
        review_comments.append(real_soup_comment)
    except :
        real_soup_comment.append('데이터없음')
review_comments
```

```python
df = pd.DataFrame(columns = ['review_score', 'review_comment'])
df['review_comment'] = review_comments
df['review_score'] = review_score
df
```
